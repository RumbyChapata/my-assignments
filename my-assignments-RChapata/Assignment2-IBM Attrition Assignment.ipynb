```python
#Importing required libraries
import pandas as pd
import numpy as np
```

```python
employees=pd.read_csv('HR-Employee-Attrition.csv')
```

```python
employees.head()
```

```python
#Checking if any columns has NA or missing values
employees.isnull().any()
```

# Showing a breakdown of distance from home by job role and attrition

```python
employees_breakdown=employees.groupby(['Attrition','JobRole'])['DistanceFromHome'].sum().reset_index()
employees_breakdown
```

# Comparing average monthly income by education and attrition

```python
inc=employees.groupby(['Attrition','Education'])['MonthlyIncome'].sum().reset_index()
inc
```

Basing on the two tables created above, distance seems to be a great factor that is contributing to employee attrition. Employees who live close to work seem to be the one staying as compared to those that live far away from work. Also the monthly income is contributing to employee attrition as employees that earn more money are staying whereas those who earn less are leaving. My recommendation is to have an incentive for those who live far from work and also review and increase monthly income so that employees can stay. Also another way to maintain a trained workforce is to employee the people who live close by. That way we can reduce the costs of training employees who wont stay for the benefit of the company.
